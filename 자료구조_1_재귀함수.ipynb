```python
def factorial(n):
  return factorial(n-1)*n

if __name__ == "__main__":
  for i in range(1,6):
    print(factorial(i))
    
```

```python
def factorial(n):
    # base case: 재귀 함수가 더 이상 호출되지 않도록 하는 ‘특정 조건’필요
    if n <= 0: # 1
        return 1
    return n*factorial(n-1)

if __name__ == "__main__":
    for i in range(1, 6):
        print(factorial(i))

# 재귀 함수를 구현하려면 1) 언제 어떤 매개변수를 가지고 호출할지 정해야 하고, 2) 호출을 정지시켜 줄 기저 사례가 필요
```

## 스택 프레임으로 재귀함수 이해

 어떻게 실행되고 있는 함수가 스스로를 다시 호출할 수 있지?
 image.png

```python
def add_two(a,b):
  c = a+b # 4
  return c

a = 10 # 1)
b = 20 # 2) 실행 시작~ 끝까지 유지되는 전역변수

result = add_two(a, b) # 3) a와 b를 인수로 전달 
                        # --> add_two 함수 호출 
                        # --> ‘스택 프레임’이라는 내부 공간 생김 
                        # --> 그 공간에 add_two 함수 내부에 있는 매개변수 a, b / 그 결괏값을 담을 지역 변수 c 저장
                        # ****# 3에서 매개변수로 전달된 a와 b는 # 1과 # 2에 있는 a와 b가 아님****
print(result) 
```

```python
# 코드 1-5 permutation.py

def permutation(arr, start): # 1) 2) 
   # 3) 기저 사례
  if len(arr)-1 == start:
    print(arr)
    return
  for idx in range(start, len(arr)): # 4) 5) 
    arr[start], arr[idx] = arr[idx], arr[start] # 6) 
    permutation(arr, start+1)
    arr[start], arr[idx] = arr[idx], arr[start] # 7) 

if __name__ == "__main__":
    arr = [1, 2, 3]
    permutation(arr, 0)

#  집합으로 쓸 리스트
#  집합의 시작 위치이며, start 이전 리스트 요소는 집합의 원소가 아닙니다.
#  기저 사례: start가 집합의 마지막 원소에 도달했을 때 섞을 다른 원소가 없으므로 완성된 순열을 출력합니다.
#  집합의 첫 번째 원소
#  idx로 집합의 모든 원소를 순회하면서 start와 idx의 원소를 바꾼 후
#  집합 크기를 줄여서, 다시 말해 start를 한 칸 움직여서 다시 재귀 함수를 호출합니다.
#  다시 자신의 스택 프레임으로 돌아왔다면 이전에 바꾸어 놓았던 원소를 다시 원래대로 돌려놓습니다.
```
